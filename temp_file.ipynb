```python
from typing import Callable
def empty_deco(func:Callable):
    # вызываем обертку
    def wrapper(): 
        # Вызов самой функции, которую нужно расширить
        res = func()
        return res
        # возвращаем новый функционал
    return wrapper
```

```python
@empty_deco
def my_func():
    return 123
print(my_func())
```

### Декоратор для подсчета времени(простой) 

```python
from typing import Callable
import time

def time_deco(func:Callable):
    # вызываем обертку
    def wrapper(): 
        start = time.time()
        # Вызов самой функции, которую нужно расширить
        res = func()
        end = time.time()
        print (f"Время выполнения заняло {end-start}")
        return res
        # возвращаем новый функционал
    return wrapper
```

```python
@time_deco
def time_funcI():
    time.sleep(3)
    return 5546
time_funcI()
```
